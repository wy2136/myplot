```python
%matplotlib inline
from myplot.geoplots import geoplot
from myplot import load_test_data
import seaborn as sns
sns.set_context('notebook')
```

Simply show a basemap:

```python
mt.geoplot()
```

The default map projection is <code>cyl</code> in the basemap package. If you want to use a different projection, e.g. <code>moll</code>, simply provide the <code>proj</code> or <code>projection</code> parameter:

```python
mt.geoplot(proj='moll')
```

Sometimes, it is better to fill the continents instead of drawing the coastlines, e.g. when you want to show ocean data. In this case, set the parameter <code>fill_continents</code> to be True.

```python
mt.geoplot(proj='moll', fill_continents=True)
```

What if we want to put the longitude 0 at the center of the base map? Just provide the <code>lon_0</code> parameter.

```python
mt.geoplot(proj='moll', fill_continents=True, lon_0=0)
```

Similarly, you can also set the `lat_0` parameter in some map projections to change the center latitude. For example, in the <code>ortho</code> projection, set the 45$^{\circ}$N at the center of the map.

```python
mt.geoplot(proj='ortho', fill_continents=True, lon_0=0, lat_0=45)
```

Now, let's show some data over the base map. First, we load some example data.

```python
zz, lon, lat = load_test_data()
```

```python
mt.geoplot(zz, lon, lat)
```

As we can see, the `matplotlib.pyplot` function `pcolormesh` is used to make the 2-D plot. You can also use a different plot function, e.g. `contourf`.

```python
mt.geoplot(zz, lon, lat, plot_type='contourf')
```

Or `contour`:

```python
mt.geoplot(zz, lon, lat, plot_type='contour')
```

Or even both `contourf` and `contour`:

```python
mt.geoplot(zz, lon, lat, plot_type='contourf+')
```

Now, let's combine different parameters described above to make more fun plots.

```python
mt.geoplot(zz, lon, lat, proj='moll', plot_type='contourf+')
```

```python
mt.geoplot(zz, lon, lat, proj='ortho', lon_0=0, lat_0=60, fill_continents=True)
```
